```python
import pandas as pd

csv_file = "AirbnbUK.csv"

#The header=0 means column names are in the first row and set the Name column to be row index
df = pd.read_csv(csv_file)

# Extract amenities column and convert to list
amenity = csv_file['amenities'].tolist()

# Create a dictionary to count the occurrence of each amenity
amenity_count = {}

for a in amenity:
    # Remove brackets and quotes
    a = a.replace('[', '').replace(']', '').replace('"', '')
    # Split amenities by comma and add to dictionary
    for amenity in a.split(','):
        amenity = amenity.strip()
        if amenity in amenity_count:
            amenity_count[amenity] += 1
        else:
            amenity_count[amenity] = 1

# Sort amenities by count in descending order
sorted_amenities = sorted(amenity_count.items(), key=lambda x: x[1], reverse=True)

# Print the top 10 amenities
for i in range(10):
    print(f"{sorted_amenities[i][0]}: {sorted_amenities[i][1]}")
```

```python
import pandas as pd

# Load Airbnb data
csv_file = pd.read_csv('AirbnbUK.csv')

# Group data by location and calculate average price
average_price_by_location = csv_file.groupby('host_location')['price'].mean().reset_index()

# Print the average price by location
print(average_price_by_location)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Airbnb data
csv_file = pd.read_csv('AirbnbUK.csv')

# Calculate the proportion of each number of bedrooms
count_of_bedrooms = csv_file['bedrooms'].value_counts()
bedroom_proportions = count_of_bedrooms / count_of_bedrooms.sum()

# Create a pie chart
labels = bedroom_proportions.index.tolist()
sizes = bedroom_proportions.tolist()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add a title
plt.title('Proportion of Number of Bedrooms')

# Show the chart
plt.show()
```
